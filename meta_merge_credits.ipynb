```python
# change str to int
new_metadata['id'] = new_metadata['id'].astype(int)
```

```python
# merge two dataframes
df = pd.merge(new_metadata, credits, on='id')
df.head(10)
```

```python
# abstract syntax tree
import ast
```

```python
# literal_eval transforms a string into a list
df['genres'] = df['genres'].map(lambda x: ast.literal_eval(x))
df['cast'] = df['cast'].map(lambda x: ast.literal_eval(x))
df['crew'] = df['crew'].map(lambda x: ast.literal_eval(x))
df.head(10)
```

```python
def create_genres_list(lis):
    genres = []
    
    for item in lis:
        if item.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            genres.append(scifi)
        else:
            genres.append(item.get('name'))
    if not genres:
        return np.NaN
    else:
        return ("".join(genres))
```

```python
df['genres_list'] = df['genres'].map(lambda x: create_genres_list(x))
```

```python
df['genres_list']
```

```python
def get_person(lis, who):
    """


    :param lis: list of strings
    :param who: actor_1_name, actor_2_name, actor_3_name
    :return: name depending on who
    """
    people = []
    for item in lis:
        people.append(item.get('name'))
    print(people)
    if who == 'actor_1_name' and len(people) > 0:
        return (people[0])
    elif who == 'actor_2_name' and len(people) > 1:
            return (people[0])
    elif who == 'actor_3_name' and len(people) > 2:
        return (people[2])
    else:
        return np.NaN

```

```python
for who in ['actor_1_name', 'actor_2_name', 'actor_3_name']:
    df[who] = df['cast'].map(lambda l: get_person(l, who))



#df['actor_1_name']
```

```python
df.columns
```

```python
def get_directors(x):
    dt = []
    
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if not dt:
        return np.NaN
    else:
        return ("".join(dt))
```

```python
df['director_name'] = df['crew'].map(lambda x: get_directors(x))
df['director_name']
```
